```python
import numpy as np
from scipy import stats

data = [12, 15, 14, 10, 18, 20, 12, 14, 15]

# Mean
mean = np.mean(data)
print("Mean:", mean)

# Median
median = np.median(data)
print("Median:", median)

# Mode
mode = stats.mode(data)
print("Mode:", mode.mode) # Access the mode value directly without indexing
#Since mode.mode may return an array if there are multiple modes,
# you could access the first mode using mode.mode[0] if needed.
# However, in this case, since there's only one mode it's a single value
# so we can just use mode.mode directly.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(data, bins=5, kde=True)
plt.axvline(mean, color='red', label='Mean')
plt.axvline(median, color='green', label='Median')
plt.legend()
plt.show()
```
