```python
from scipy.stats import f_oneway

# Sample groups
group1 = [23, 20, 22, 21]
group2 = [27, 29, 28, 26]
group3 = [31, 33, 30, 29]

# One-way ANOVA
f_stat, p_value = f_oneway(group1, group2, group3)

print("F-statistic:", f_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Significant difference among groups.")
else:
    print("No significant difference among groups.")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

group1 = [23, 20, 22, 21]
group2 = [27, 29, 28, 26]
group3 = [31, 33, 30, 29]

df = pd.DataFrame({
    "Score": group1 + group2 + group3,
    "Group": ["Group 1"]*len(group1) + ["Group 2"]*len(group2) + ["Group 3"]*len(group3)
})

sns.boxplot(x="Group", y="Score", data=df)
plt.title("ANOVA Comparison Across Groups")
plt.show()
```
