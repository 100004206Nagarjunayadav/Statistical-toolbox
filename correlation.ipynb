```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]

# Pearson correlation
corr = np.corrcoef(x, y)[0, 1]
print("Pearson Correlation Coefficient:", corr)

# Visualization
sns.scatterplot(x=x, y=y)
plt.title(f"Correlation (r = {corr:.2f})")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
